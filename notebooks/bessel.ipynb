```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.special import jv
```

```python
fig, ax = plt.subplots()
x = np.linspace(-10., 10., 1000)
for i in range(4):
    ax.plot(x, jv(i, x), label=f'$J_{i!r}$')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
x = np.linspace(-15., 15., 1000)
y = jv(0, x)
ax.plot(x, y)
ax.legend()
plt.show()
```
